```python
import numpy as np
import pandas as pd
from sklearn.metrics import r2_score,mean_squared_error
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
path='MD1/csv/'
model='out_step1.csv'
params='param_step1.csv'
df_model = pd.read_csv(path+model,header=None)
df_param = pd.read_csv(path+params,header=None)
```

```python
model='out_step2.csv'
params='param_step2.csv'
df_model = pd.read_csv(path+model,header=None)
df_param = pd.read_csv(path+params,header=None)
```

```python
def match_plot(df,logy=False):
    df.iloc[0:-1,:].transpose().plot(logy=logy,legend=False,alpha=0.5,figsize=(10,5))
    df.iloc[-1,:].plot(logy=logy,legend=False,style="o",color='red');

match_plot(df_model)
match_plot(df_model,logy=True)
```

```python
[n,m]=np.shape(df_model)
r2=[r2_score(df_model.iloc[i,:], df_model.iloc[-1,:]) for i in range(n-1)]
rmse=[mean_squared_error(df_model.iloc[i,:], df_model.iloc[-1,:]) for i in range(n-1)]
r2=np.asarray(r2)
rmse=np.asarray(rmse)
plt.plot(rmse,r2,'o'), plt.xlabel('RMSE'), plt.ylabel('$R^2$');
```

```python
tight_params=df_param[r2>0.955]
tight_model=df_model.iloc[0:-1,:][r2>0.955]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
#ax1.plot(x, y)
ax1.plot(df_param.iloc[:,0],df_model.iloc[0:-1,0],'o',alpha=0.5,color='b')
ax1.set_xlabel('nmax0')
ax1.set_ylabel('NPP')
ax1.set_ylim([min(df_model.iloc[:,0])-1, max(df_model.iloc[:,0]+1)])

ax1.plot(np.linspace(1,5,10),np.ones(10)*df_model.iloc[-1,0],alpha=0.5,color='black')
#ax2.plot(x, -y)

ax2.plot(tight_params.iloc[:,0],tight_model.iloc[:,0],'o',alpha=0.5,color='b')
ax1.set_xlabel('nmax0')
ax2.plot(np.linspace(1,5,10),np.ones(10)*df_model.iloc[-1,0],alpha=0.5,color='black')
ax2.set_ylim([min(df_model.iloc[:,0])-1, max(df_model.iloc[:,0])+1])
```
